```python
# Load data
df = pd.read_csv('train.csv')
```

```python
# 1. DATA OVERVIEW
print(df.info())          # Data types, missing values
print(df.describe(include='all'))   # Statistical summary
print(df.head())          # First 5 rows for quick glance
```

```python
# 2. CATEGORICAL FEATURE DISTRIBUTIONS
print("Sex value counts:")
print(df['Sex'].value_counts())
print("Pclass value counts:")
print(df['Pclass'].value_counts())
print("Embarked value counts:")
print(df['Embarked'].value_counts())
print("Missing values in each column:")
print(df.isnull().sum())
```

```python
# 3. HISTOGRAMS FOR NUMERICAL COLUMNS
num_cols = ['Age', 'Fare', 'SibSp', 'Parch']
df[num_cols].hist(bins=20, figsize=(12,8))
plt.suptitle('Histogram Distribution of Numerical Features')
plt.tight_layout()
plt.show()
```

```python
# 4. BOXPLOTS FOR OUTLIER DETECTION
for col in num_cols:
    plt.figure()
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot of {col}')
    plt.show()
```

```python
# 5. PAIRPLOT & CORRELATION HEATMAP
sns.pairplot(df[['Survived', 'Age', 'Fare', 'SibSp', 'Parch']].dropna())
plt.suptitle('Pairplot of Key Numerical Features')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
corr = df[['Survived', 'Age', 'Fare', 'SibSp', 'Parch']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# 6. SCATTERPLOT EXAMPLE: AGE vs. FARE (Remove NaNs)
sns.scatterplot(data=df, x='Age', y='Fare', hue='Survived')
plt.title('Age vs Fare by Survival Status')
plt.show()
```

Key Observations & Insights
Data Structure & Missing Values
Several columns have missing values, especially Age and Cabin.

Survived has a balanced 0/1 outcome (saved or lost).

Most passengers are male (577 vs. 314 female).

Univariate Distributions
Age: Passengers are mostly young adults (mean29.7), but ages range widely (infants to seniors).

Fare: Most fares are low, but there are a few extremely high (skewed right, some outliers).

SibSp/Parch: Majority traveled alone or with one family member; large family groups are rare.

Category Distributions
Pclass: Most passengers are in 3rd class (low fare, high risk), fewer in 1st class.

Embarked: S (Southampton) is the most common port, followed by C (Cherbourg) and Q (Queenstown).

Outlier Detection
Boxplots clearly show extreme outliers in Fare, a few in Age (very old), and large family groups (SibSp/Parch).

Relationships & Trends
Heatmap: There are weak correlations among features; for EDA, none exceed |0.5|.

Pairplot: No strong linear relationships, but higher fare often means higher class and greater survival.

Scatterplot (Age vs Fare): Elderly and children appear in both high/low fare categories.

Survival Insights
Survival (binary) is more likely for:

Females, children, and first-class ticket holders.

Those with higher fares.

Summary of Findings
The dataset is rich with categorical and continuous features, but many columns have missing values, especially Age and Cabin.

Most people traveled in third class, with significant survival disparities across Sex, Pclass, and Fare.

High fare and first class correlate with increased survival. Females and children also have better odds.

There are notable outliers in fare (luxury tickets) and family groups, which may require cleaning for modeling.

No strong linear correlations among numerical features, but combinations (e.g. young age + high fare) are indicative.

Embarked port distribution is uneven, with Southampton dominating. Some missing data points here.
